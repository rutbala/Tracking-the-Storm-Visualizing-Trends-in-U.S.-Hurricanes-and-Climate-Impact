```python
# Converting the 'Date' column to a datetime object for easier manipulation
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'], format='%Y%m%d', errors='coerce')
```

```python
# Extracting year and month from the Date column
data_cleaned['Year'] = data_cleaned['Date'].dt.year
data_cleaned['Month'] = data_cleaned['Date'].dt.month
```

```python
def convert_lat_lon(value):
    """Convert latitude and longitude strings to numeric values."""
    try:
        value = str(value)  # Ensure the value is a string
        if value[-1] in ['N', 'E']:
            return float(value[:-1])  # Positive for North and East
        elif value[-1] in ['S', 'W']:
            return -float(value[:-1])  # Negative for South and West
        else:
            return float(value)  # If already numeric
    except (ValueError, IndexError):
        return None  # Return None for invalid cases
```

```python
# Apply the conversion function to Latitude and Longitude
data_cleaned['Latitude'] = data_cleaned['Latitude'].apply(convert_lat_lon)
data_cleaned['Longitude'] = data_cleaned['Longitude'].apply(convert_lat_lon)

# Drop rows with invalid Latitude or Longitude values
data_cleaned = data_cleaned.dropna(subset=['Latitude', 'Longitude'])
```

**17.Hurricane Direction Distribution**

```python
# Sort data by ID and Date
data = data.sort_values(by=['ID', 'Date', 'Time'])

# Calculate differences in latitude and longitude
data['Lat_diff'] = data.groupby('ID')['Latitude'].diff()
data['Lon_diff'] = data.groupby('ID')['Longitude'].diff()

# Calculate bearing (angle of movement in degrees)
data['Bearing'] = np.degrees(np.arctan2(data['Lon_diff'], data['Lat_diff']))
data['Bearing'] = (data['Bearing'] + 360) % 360

# Bin bearings into cardinal directions
bins = [0, 45, 90, 135, 180, 225, 270, 315, 360]
labels = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']
data['Direction'] = pd.cut(data['Bearing'], bins=bins, labels=labels, include_lowest=True)

# Count hurricanes in each direction
direction_counts = data['Direction'].value_counts().reindex(labels)

# Plot the distribution on a polar bar plot
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
counts = direction_counts.tolist()

# Close the circle for plotting
angles += angles[:1]
counts += counts[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})
ax.fill(angles, counts, color='blue', alpha=0.3)
ax.bar(angles, counts, color='blue', alpha=0.7, width=0.4)

# Add labels for cardinal directions
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

ax.set_title("Hurricane Direction Distribution", va='bottom')
plt.show()
```

**Highlights**

* This polar chart illustrates the distribution of hurricane directions, showing the frequency of storms originating or moving in specific compass directions.

* The most prominent activity is concentrated in the northeast (NE) quadrant, indicating that hurricanes predominantly move in this direction, likely influenced by atmospheric pressure systems and prevailing winds.

* The northwest (NW) and east (E) directions also exhibit moderate activity, while the southern quadrants (SW and SE) show minimal frequency. The radial values quantify the number of hurricanes for each direction, highlighting the dominant directional patterns.

**19. Distribution of Hurricane Lifespans**

```python
# Ensure the 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')

# Group by hurricane ID to calculate lifespan
hurricane_lifespans = data.groupby('ID').agg(
    start_date=('Date', 'min'),
    end_date=('Date', 'max')
)
hurricane_lifespans['Lifespan'] = (hurricane_lifespans['end_date'] - hurricane_lifespans['start_date']).dt.days

# Filter non-positive lifespans
hurricane_lifespans = hurricane_lifespans[hurricane_lifespans['Lifespan'] > 0]

# Plot Histogram of Lifespan
plt.figure(figsize=(12, 6))
plt.hist(hurricane_lifespans['Lifespan'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Hurricane Lifespans')
plt.xlabel('Lifespan (Days)')
plt.ylabel('Number of Hurricanes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Highlights**

* This histogram displays the distribution of hurricane lifespans measured in days. The majority of hurricanes last between 1 to 5 days, with the highest frequency occurring in this range.

* Beyond 5 days, the number of hurricanes decreases steadily, indicating that fewer storms persist for longer durations.

**20. Average Wind Speeds by  Directions**

```python
# Adjust to use only the directions available in the dataset: NE, SE, SW, NW
available_directions = ['NE', 'SE', 'SW', 'NW']

# Extract relevant wind data for the available directions
wind_data_adjusted = {
    direction: {
        'Low Wind': data[f'Low Wind {direction}'].replace(-999, None).mean(),
        'Moderate Wind': data[f'Moderate Wind {direction}'].replace(-999, None).mean(),
        'High Wind': data[f'High Wind {direction}'].replace(-999, None).mean()
    }
    for direction in available_directions
}

# Convert the adjusted wind data into a DataFrame for plotting
wind_df_adjusted = pd.DataFrame(wind_data_adjusted).T

# Plot the wind patterns for the available directions
wind_df_adjusted.plot(kind='bar', figsize=(12, 8), alpha=0.8, color=['skyblue', 'orange', 'red'])
plt.title('Average Wind Speeds by Available Directions and Category', fontsize=14)
plt.xlabel('Direction', fontsize=12)
plt.ylabel('Average Wind Speed (knots)', fontsize=12)
plt.legend(['Low Wind', 'Moderate Wind', 'High Wind'], title='Wind Category')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.show()
```
